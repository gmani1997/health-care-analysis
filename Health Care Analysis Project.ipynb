# Data Preprocessing and Feature Engineering
### 1. Finding out the missing values and replacing them with the mean or median of the column

```python
data.info()
```

We can clearly see that data set having the null values in the Bed_Grade column and the City_Code_Patient column. Both of them are float type datas so we can easily fill those null values with the mean of the particular columns.

```python
#we have the mean values of those particular columns now lets fill the null values with those means

data.Bed_Grade.fillna(data.Bed_Grade.mean(), inplace = True)

data.City_Code_Patient.fillna(data.City_Code_Patient.mean(), inplace = True)
```

```python
#Now lets check for the null values once again

data.isnull().sum()
```

We can clearly see that we have remove all the null values from the dataset.
